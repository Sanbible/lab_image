```python
import numpy as np
from skimage import data 
from skimage import color
from skimage import filters
import matplotlib.pyplot as plt
from skimage.morphology import rectangle
```

```python
# Load an image from SKIMAGE of scikit-image
image_moon = data.moon() # read the image
print(image_moon.size)  # Array image size
print(image_moon.shape) # Width and Height of image

plt.figure(figsize=(10,3))
plt.imshow(image_moon, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')
plt.show()

# Load an image from SKIMAGE of scikit-image & fix size
image_hubble = data.hubble_deep_field()[0:500, 0:500]
print(image_hubble.size)  # Array image size
print(image_hubble.shape) # Width and Height of image

plt.figure(figsize=(10,3))
plt.imshow(image_hubble, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')
plt.show()

# Convert rgb to grayscale
image_star = color.rgb2gray(image_hubble)
print(image_star.size)  # Array image size
print(image_star.shape) # Width and Height of image

plt.figure(figsize=(10,3))
plt.imshow(image_star, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
mask_w = 3
image_avg_3 = filters.rank.mean(image_star, rectangle(mask_w, mask_w))

mask_w = 15
image_avg_15 = filters.rank.mean(image_star, rectangle(mask_w, mask_w))

mask_w = 35
image_avg_35 = filters.rank.mean(image_star, rectangle(mask_w, mask_w))

mask_w = 71
image_avg_71 = filters.rank.mean(image_star, rectangle(mask_w, mask_w))

# display the image
fig, ax = plt.subplots(ncols=5, nrows=1, 
                               figsize=(15, 4),
                               sharex=True, 
                               sharey=True)

ax[0].imshow(image_star, cmap='gray')
ax[0].title.set_text('Original Image')

ax[1].imshow(image_avg_3, cmap='gray')
ax[1].title.set_text('Avg_filter 3')

ax[2].imshow(image_avg_15, cmap='gray')
ax[2].title.set_text('Avg_filter 15')

ax[3].imshow(image_avg_35, cmap='gray')
ax[3].title.set_text('Avg_filter 35')

ax[4].imshow(image_avg_71, cmap='gray')
ax[4].title.set_text('Avg_filter 71')

for aa in ax:
    aa.set_axis_off()
plt.show()
```

```python
# Create random noise in image
noise = np.random.random(image_moon.shape)
image_noise = image_moon.copy()
image_noise[noise < 0.01] = 0 # Create pepper Noise

# display the image
plt.figure(figsize=(10,3))

plt.subplot(1,3,1)
plt.imshow(image_moon, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(image_noise, cmap='gray') # display the image
plt.title('Pepper Noise Image')
plt.axis('off')

# Maximum filter 
mask_w = 5
# image_denoise = filters.rank.maximum(image_noise, disk(mask_w))
image_denoise = filters.rank.maximum(image_noise, rectangle(mask_w, mask_w))

plt.subplot(1,3,3)
plt.imshow(image_denoise, cmap='gray') # display the image
plt.title('Reduce Pepper Noise Image')
plt.axis('off')


plt.show()
```

```python
# Create random noise in image
noise = np.random.random(image_moon.shape)
image_noise = image_moon.copy()
image_noise[noise > 0.99] = 255 # Create Salt Noise

# display the image
plt.figure(figsize=(10,3))

plt.subplot(1,3,1)
plt.imshow(image_moon, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(image_noise, cmap='gray') # display the image
plt.title('Salt Noise Image')
plt.axis('off')

# Minimum filter 
mask_w = 5
# image_denoise = filters.rank.minimum(image_noise, disk(mask_w))
image_denoise = filters.rank.minimum(image_noise, rectangle(mask_w, mask_w))

plt.subplot(1,3,3)
plt.imshow(image_denoise, cmap='gray') # display the image
plt.title('Reduce Slat Noise Image')
plt.axis('off')

plt.show()
```

```python
# Create random noise in image
noise = np.random.random(image_moon.shape)
image_noise = image_moon.copy()
image_noise[noise > 0.99] = 255 # Create Salt Noise
image_noise[noise < 0.01] = 0 # Create pepper Noise

# display the image
plt.figure(figsize=(10,10))

plt.subplot(3,2,1)
plt.imshow(image_moon, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')

plt.subplot(3,2,2)
plt.imshow(image_noise, cmap='gray') # display the image
plt.title('Salt&Pepper Noise Image')
plt.axis('off')

# Minimum filter 
mask_w = 5
# image_denoise = filters.rank.minimum(image_noise, disk(mask_w))
image_denoise = filters.rank.minimum(image_noise, rectangle(mask_w, mask_w))

plt.subplot(3,2,3)
plt.imshow(image_denoise, cmap='gray') # display the image
plt.title('Noise reduction by min')
plt.axis('off')
# Maximum filter 
mask_w = 5
image_denoise = filters.rank.maximum(image_noise, rectangle(mask_w, mask_w))

plt.subplot(3,2,4)
plt.imshow(image_denoise, cmap='gray') # display the image
plt.title('Noise reduction by max')
plt.axis('off')

# Average filter 
mask_w = 5
# image_denoise = filters.rank.median(image_noise, disk(mask_w))
image_denoise = filters.rank.mean(image_noise, rectangle(mask_w, mask_w))

plt.subplot(3,2,5)
plt.imshow(image_denoise, cmap='gray') # display the image
plt.title('Noise reduction by avg')
plt.axis('off')

# Median filter 
mask_w = 5
# image_denoise = filters.rank.median(image_noise, disk(mask_w))
image_denoise = filters.rank.median(image_noise, rectangle(mask_w, mask_w))

plt.subplot(3,2,6)
plt.imshow(image_denoise, cmap='gray') # display the image
plt.title('Noise reduction by median')
plt.axis('off')

plt.show()
```
