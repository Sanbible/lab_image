```python
import numpy as np
from skimage import data 
from skimage import exposure
import matplotlib.pyplot as plt
```

```python
# Load an image from SKIMAGE of scikit-image
image_cam = data.camera() # read the image

print(type(image_cam)) # Image in array format 
print(image_cam.size)  # Array image size
print(image_cam.shape) # Width and Height of image



# display the image
plt.figure(figsize=(10,3))

plt.subplot(1,2,1)
plt.imshow(image_cam, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')

# Load an image from SKIMAGE of scikit-image
image_moon = data.moon() # read the image

plt.subplot(1,2,2)
plt.imshow(image_moon, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')

plt.show()
```

```python
# Calculation Histogram and Cumulative Histogram
image_hist,bins = np.histogram(image_cam.flatten(),256,[0,256])
image_cdf = image_hist.cumsum()
cdf_normalized = image_cdf * image_hist.max()/ image_cdf.max()


# display the image
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(image_cam, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.hist(image_cam.ravel(),256,[0,256]);
# plt.xlabel('Pixel intensity')
plt.title('Histogram')
  

plt.subplot(2,2,3)
plt.plot(cdf_normalized, color = 'r');
# plt.xlabel('Pixel intensity')
plt.title('Cumulative Histogram')


plt.subplot(2,2,4)
plt.plot(cdf_normalized, color = 'r')
plt.hist(image_cam.flatten(),256,[0,256])
# plt.xlabel('Pixel intensity')
plt.legend(('CDF','Hist'), loc = 'center right')
plt.title('Histogram vs Cumulative')

plt.show()
```

```python
# Load an image from SKIMAGE of scikit-image
image_cam = data.camera() # read the image

# display the image
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(image_cam, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.hist(image_cam.ravel(),256,[0,256]);
# plt.xlabel('Pixel intensity')
plt.title('Histogram')
  
# inverted_img = util.invert(image)
imgage_inverse = 255 - image_cam

plt.subplot(2,2,3)
plt.imshow(imgage_inverse, cmap='gray') # display the image
plt.title('Inverse Image')
plt.axis('off')


plt.subplot(2,2,4)
plt.hist(imgage_inverse.ravel(),256,[0,256]);
plt.title('Inverse Histogram')

plt.show()
```

```python
# Global Histogram Equalization
image_moon_histequal = exposure.equalize_hist(image_moon)

# display the image
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(image_moon, cmap='gray') # display the image
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.hist(image_moon.ravel(),256,[0,256]);
# plt.hist(image_moon.ravel(), 256, histtype='step');
plt.title('Histogram Input')

plt.subplot(2,2,4)
plt.hist(image_moon_histequal.ravel(), 256, histtype='step');
plt.title('Histogram Output')

plt.subplot(2,2,3)
plt.imshow(image_moon_histequal, cmap='gray') # display the image
plt.title('Output Image')
plt.axis('off')

plt.show()
```
