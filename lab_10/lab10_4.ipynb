```python
import cv2
import os
import matplotlib.pyplot as plt

# Load the cascade with full path
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Check if cascade loaded successfully
if face_cascade.empty():
    print("Error loading cascade file")
    exit()

# Read the input image
img = cv2.imread('Lenna_(test_image).png')

# Check if image loaded successfully
if img is None:
    print("Error loading image file")
    exit()

# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Adjust the histogram
gray = cv2.equalizeHist(gray)

# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

print(f"Found {len(faces)} faces")

# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Convert BGR to RGB for displaying with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image below the code cell
plt.imshow(img_rgb)
plt.axis('off')
plt.title('Face Detection')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Read the input image
img = cv2.imread('Lenna_(test_image).png')

# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Adjust the histogram
gray = cv2.equalizeHist(gray)

# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    faceROI = gray[y:y+h, x:x+w]
    # In each face, detect eyes
    eyes = eye_cascade.detectMultiScale(faceROI, 1.1, 8)
    for (x2, y2, w2, h2) in eyes:
        eye_center = (x + x2 + w2//2, y + y2 + h2//2)
        radius = int(round((w2 + h2)*0.25))
        cv2.circle(img, eye_center, radius, (0, 0, 255), 2)

# Convert BGR to RGB for displaying with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image below the code cell
plt.imshow(img_rgb)
plt.axis('off')
plt.title('Face and Eyes Detection')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Read the input image
img = cv2.imread('IMG-3006_Cs4r.jpg')

# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.equalizeHist(gray)

# Adjust the parameters of Detect faces
faces = face_cascade.detectMultiScale(gray, 1.005, 2, minSize=(31,31), maxSize=(70,70))

# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Convert BGR to RGB for displaying with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image below the code cell
plt.imshow(img_rgb)
plt.axis('off')
plt.title('Face Detection')
plt.show()
```
