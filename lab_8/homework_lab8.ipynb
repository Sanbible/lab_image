```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
print(kernel)
```

```python
img = cv2.imread("HW_fingerprint.png") 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow( gray, cmap = 'gray')
plt.show()
```

```python
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
print(binary)
```

```python
print("binary dtype:", binary.dtype, "shape:", binary.shape)
```

```python
erosion1 = cv2.erode(binary, kernel, iterations=1)
plt.imshow( erosion1 , cmap= 'gray')
plt.show()
```

```python
dilation1 = cv2.dilate(erosion1, kernel, iterations=1)
plt.imshow( dilation1 , cmap= 'gray')
plt.show()
```

```python
dilation2 = cv2.dilate(dilation1, kernel, iterations=1)
plt.imshow( dilation2 , cmap= 'gray')
plt.show()
```

```python
erosion2 = cv2.erode(dilation2, kernel, iterations=1)
plt.imshow( erosion2 , cmap= 'gray')
plt.show()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(12,4))

axes[0].imshow(img, cmap = 'gray')
axes[0].set_title("Original Picture")
axes[0].axis('off')

axes[1].imshow(erosion1, cmap='gray')
axes[1].set_title("Erosion (1st)")
axes[1].axis('off')

axes[2].imshow(dilation1, cmap='gray')
axes[2].set_title("Dilation (1st)")
axes[2].axis('off')

axes[3].imshow(dilation2, cmap='gray')
axes[3].set_title("Dilation (2nd)")
axes[3].axis('off')

axes[4].imshow(erosion2, cmap='gray')
axes[4].set_title("Erosion (final)")
axes[4].axis('off')


plt.show()
```
