```python
from skimage.filters import threshold_minimum
from skimage.filters import threshold_mean
from skimage import data
import matplotlib.pyplot as plt

image = data.camera()
```

```python
thresh_min = threshold_minimum(image)
binary_min = image > thresh_min
```

```python

plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.imshow(image, cmap=plt.cm.gray)
plt.title('Original')

plt.subplot(222)
plt.hist(image.ravel(), bins=256)
plt.title('Histogram')

plt.subplot(224)
plt.hist(image.ravel(), bins=256)
plt.axvline(thresh_min, color='r')
plt.title('Histogram selection min = %d' %(thresh_min))

plt.subplot(223)
plt.imshow(binary_min, cmap=plt.cm.gray)
plt.title('Thresholded min Output')
    
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
thresh_mean = threshold_mean(image)
binary_mean = image > thresh_mean
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.imshow(image, cmap=plt.cm.gray)
plt.title('Original')

plt.subplot(222)
plt.hist(image.ravel(), bins=256)
plt.title('Histogram')

plt.subplot(224)
plt.hist(image.ravel(), bins=256)
plt.axvline(thresh_mean, color='r')
plt.title('Histogram selection mean = %d' %(thresh_mean))

plt.subplot(223)
plt.imshow(binary_mean, cmap=plt.cm.gray)
plt.title('Thresholded mean Output')
    
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
image = data.page()

plt.figure(figsize=(15, 15))

# Original
plt.subplot(421)
plt.imshow(image, cmap=plt.cm.gray)
plt.title('Original')

plt.subplot(422)
plt.hist(image.ravel(), bins=256)
plt.title('Histogram')

# Minimum
thresh_min = threshold_minimum(image)
binary_min = image > thresh_min

plt.subplot(424)
plt.hist(image.ravel(), bins=256)
plt.axvline(thresh_min, color='r')
plt.title('Histogram selection = %d' %(thresh_min))

plt.subplot(423)
plt.imshow(binary_min, cmap=plt.cm.gray)
plt.title('Thresholded Output')

# Mean
thresh_mean = threshold_mean(image)
binary_mean = image > thresh_mean

plt.subplot(426)
plt.hist(image.ravel(), bins=256)
plt.axvline(thresh_mean, color='r')
plt.title('Histogram selection = %d' %(thresh_mean))

plt.subplot(425)
plt.imshow(binary_mean, cmap=plt.cm.gray)
plt.title('Thresholded Output')

# Selection
thresh_select = 137
binary_select = image > thresh_select

plt.subplot(428)
plt.hist(image.ravel(), bins=256)
plt.axvline(thresh_select, color='r')
plt.title('Histogram selection = %d' %(thresh_select))

plt.subplot(427)
plt.imshow(binary_select, cmap=plt.cm.gray)
plt.title('Thresholded Output')
    
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from skimage.morphology import disk
from skimage.filters import threshold_otsu, rank
from skimage.util import img_as_ubyte

image = img_as_ubyte(data.page())

plt.figure(figsize=(15, 15))

plt.subplot(421)
plt.imshow(image, cmap=plt.cm.gray)
plt.title('Original')
plt.axis('off')


# Small
radius = 5
selem = disk(radius)

local_otsu = rank.otsu(image, selem)
local_otsu_output = image >= local_otsu

plt.subplot(423)
plt.imshow(local_otsu, cmap=plt.cm.gray)
plt.title('Local Otsu (radius=%d)' % radius)
plt.axis('off')

plt.subplot(424)
plt.imshow(local_otsu_output, cmap=plt.cm.gray)
plt.title('Local Otsu' % local_otsu)
plt.axis('off')

# Big
radius = 20
selem = disk(radius)

local_otsu = rank.otsu(image, selem)
local_otsu_output = image >= local_otsu

plt.subplot(425)
plt.imshow(local_otsu, cmap=plt.cm.gray)
plt.title('Local Otsu (radius=%d)' % radius)
plt.axis('off')

plt.subplot(426)
plt.imshow(local_otsu_output, cmap=plt.cm.gray)
plt.title('Local Otsu' % local_otsu)
plt.axis('off')

# Large
radius = 40
selem = disk(radius)

local_otsu = rank.otsu(image, selem)
local_otsu_output = image >= local_otsu

plt.subplot(427)
plt.imshow(local_otsu, cmap=plt.cm.gray)
plt.title('Local Otsu (radius=%d)' % radius)
plt.axis('off')

plt.subplot(428)
plt.imshow(local_otsu_output, cmap=plt.cm.gray)
plt.title('Local Otsu' % local_otsu)
plt.axis('off')

plt.xticks(())
plt.yticks(())
plt.show()
```
